```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def dydt(y,t):
    return 5*y-5*t**(2)
```

```python
def f(t):
    return t**(2)+(2/5)*t+(2/25)
```

```python
n = 2500
t = np.linspace(0,2,n)
h = t[1]-t[0]
print(h)
```

```python
plt.plot(t,f(t))
```

```python

def euler(y,t):
    y = np.zeros(len(t))
    y[0] = 2/25
    #plt.plot(t,f(t))
    #plt.show()
    for i in range(len(t)-1):
        y[i+1] = y[i]+h*dydt(y[i],t[i])
    return y


    

    
plt.plot(t,euler((2/25),t),'g')
```

```python
def RK2(y,t):
    y = np.zeros(len(t))
    y[0] = 2/25
    #plt.plot(t,f(t))
    #plt.show()
    for i in range(len(t)-1):
        y_step = y[i]+h*dydt(y[i],t[i])
        y[i+1] = y[i]+(h/2)*(dydt(y[i],t[i])+dydt(y_step,t[i+1]))
        
    return y   
                                                                   

plt.plot(t,RK2((2/25),t),'r',label='RK2')  
plt.plot(t,euler((2/25),t),'g',label='Euler')
plt.legend()
```

```python
def Imp_euler(y0,t):
    y0 = (2/25)
    y = np.zeros(len(t))
    y[0] = y0
    for i in range(len(t)-1):
        y[i+1] = (y[i]-h*5*(t[i+1]**(2)))/((1-h*5))
        
    return y    
plt.figure(figsize=(14,7),dpi=120)
plt.plot(t,f(t),'b-',label='Analytical')
plt.plot(t,Imp_euler((2/25),t),'k--',label='Imp_euler')
plt.plot(t,RK2((2/25),t),'r--',label='RK2')  
plt.plot(t,euler((2/25),t),'g--',label='Forward_euler')

plt.legend()
plt.grid()

```
